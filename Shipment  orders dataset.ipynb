### Which sub category had highest growth by profit in 2023 compare to 2022

```python
m = df1.groupby(['Sub Category', 'Year']).sum(['Total Profit'])['Total Profit'].reset_index()
m
```

```python
n = m.pivot_table(index = 'Sub Category', columns='Year', values = 'Total Profit').reset_index()
n.head()
```

```python
n['diff'] = n['01-01-2023'] - n['31-12-2022']

n.sort_values(by = ['diff'], ascending=False).head(5)
```
